```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
%matplotlib inline
```

### Total Column Water Vapor

```python
f = '/pscratch/sd/s/smheflin/pnw_mcs/e5.oper.an.sfc.128_137_tcwv.ll025sc.2024080100_2024083123.nc'
ds = xr.open_dataset(f)
lat = ds.latitude
lon = ds.longitude
```

```python
data = ds.sel(time = '2024-08-17T02:00:00.000000000').TCWV
fig = plt.figure(dpi = 200)
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.set_extent([-160, -80,20, 70], crs=ccrs.PlateCarree())
levels = np.linspace(0,81,10) 
a = ax.contourf(lon, lat, data, levels = levels)
ax.add_feature(cfeature.COASTLINE)
proj = ccrs.PlateCarree()
gl = ax.gridlines(crs=proj, draw_labels=True, linestyle='--', linewidth=0.5)
ax.set_title(str(data.time.data)[:19])
gl.right_labels = False
gl.top_labels = False
ax.add_feature(cfeature.STATES, edgecolor='k', linewidth=0.5)
plt.colorbar(a, ax = ax,fraction = 0.08, pad = 0.1, label = r'Precipitable Water ($kg/m^2$)', orientation = 'horizontal', extend='both')
```

### Integrated vapor transport

```python
f = '/pscratch/sd/s/smheflin/pnw_mcs/e5.oper.an.vinteg.162_071_viwve.ll025sc.2024080100_2024083123.nc'
ds_x = xr.open_dataset(f).VIWVE.sel(time = '2024-08-17T18:00:00.000000000')
f = '/pscratch/sd/s/smheflin/pnw_mcs/e5.oper.an.vinteg.162_072_viwvn.ll025sc.2024080100_2024083123.nc'
ds_y = xr.open_dataset(f).VIWVN.sel(time = '2024-08-17T18:00:00.000000000')
```

```python
mag = np.sqrt(ds_x.data**2 + ds_y.data**2)
```

```python
##data = ds.sel(time = '2024-08-17T02:00:00.000000000').TCWV
fig = plt.figure(dpi = 200)
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.set_extent([-160, -80,20, 70], crs=ccrs.PlateCarree())
# = np.linspace(0,81,10) 
a = ax.pcolormesh(lon, lat, mag, cmap = 'plasma', vmax = 1000) # levels = levels)
ax.add_feature(cfeature.COASTLINE)
proj = ccrs.PlateCarree()
gl = ax.gridlines(crs=proj, draw_labels=True, linestyle='--', linewidth=0.5)
ax.set_title(str(ds_x.time.data)[:19])
gl.right_labels = False
gl.top_labels = False
ax.add_feature(cfeature.STATES, edgecolor='k', linewidth=0.5)
plt.colorbar(a, ax = ax,fraction = 0.08, pad = 0.1, label = r'Integrated Vapor Flux ($kg/m/s$)', orientation = 'horizontal', extend='both')
```

### Sea surface temperature

```python
f = '/pscratch/sd/s/smheflin/pnw_mcs/e5.oper.an.sfc.128_034_sstk.ll025sc.2024080100_2024083123.nc'
ds = xr.open_dataset(f)
data = ds.SSTK.sel(time = '2024-08-17T03:00:00.000000000') - 273.15
lon = data.longitude
lat = data.latitude
fig = plt.figure(dpi = 200)

ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.set_extent([-160+365, -100+365,20, 70], crs=ccrs.PlateCarree())
a = ax.pcolormesh(lon, lat, data, cmap = 'coolwarm', vmin = 0)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.LAKES)

proj = ccrs.PlateCarree()
gl = ax.gridlines(crs=proj, draw_labels=True, linestyle='--', linewidth=0.5)
ax.set_title(str(ds_x.time.data)[:19])
gl.right_labels = False
gl.top_labels = False

ax.add_feature(cfeature.STATES, edgecolor='k', linewidth=0.5)
plt.colorbar(a, ax = ax,fraction = 0.06, pad = 0.1, label = r'Sea Surface Temperature (K)', orientation = 'horizontal', extend='both')
```
